```r
library(dplyr)
library(readr)
library(ggplot2)
library(ggthemes)
```

```r
df <- read_csv("https://data.novascotia.ca/api/views/55ax-8cpp/rows.csv?accessType=DOWNLOAD")
```

```r
glimpse(df)
```

```r
names(df) <- c("Years_averaged", "Sex", "Age_Group", "Mean_annual_frequency", "Mean_rate_per_100K_population")
```

```r
df
```

```r
df$Age_Group <- factor(df$Age_Group)
```

```r
df$Sex <- factor(df$Sex)
```

```r
df$Years_averaged <- factor(df$Years_averaged)
```

```r
df2013_2016 <- df %>%
  filter(Years_averaged == "2013-2016")
```

```r
glimpse(df2013_2016)
```

```r
df2013_2016
```

```r
myplot <- ggplot(df2013_2016, aes(x = Sex, y = Mean_annual_frequency, color = Sex)) +
  geom_point() +
  facet_wrap(~ Age_Group) +
  ylab("Mean number of suicides") +
  xlab("Sex by Age Group") +
  ggtitle("Nova Scotia Average Yearly Suicides\nby Sex and Age from 2013-2016") +
  theme_solarized()
```

```r
myplot
```
