# Storytelling with Open Data and Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sys

print(sys.version)
```

#### About this dataset: https://data.novascotia.ca/Health-and-Wellness/Lab-confirmed-Influenza-Cases/7af8-mg86

```python
df = pd.read_csv("https://data.novascotia.ca/api/views/7af8-mg86/rows.csv?accessType=DOWNLOAD")
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Week Ending'] = df['Week Ending'].str.split().str.get(0)
```

```python
df_2016 = df[df['Week Ending'].str.contains("2016")]
```

```python
df_2016.info()
```

```python
plt.rcParams['figure.figsize']=(8,16)
```

```python
plt.figure()
```

```python
sns.set(font_scale=1.25)
```

```python
ax = sns.barplot(y="Week Ending", x="Lab-Confirmed Influenza Cases", data=df_2016, palette="Blues_d")
ax.set_title('Nova Scotia Lab-Confirmed Influenza Cases 2016')
plt.ylabel('Week Ending')
plt.xlabel('Count')
```
